```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

### Model 12
plt.suptitle('Classifer 2, optimal parameters from papers', fontsize=20)

ax[0].plot(model12_hist['acc'])
ax[0].plot(model12_hist['val_acc'])
ax[0].set_title('model 12 accuracy')
ax[0].set_ylabel('accuracy')
ax[0].set_xlabel('epoch')
ax[0].legend(['train', 'test'], loc='lower right')

ax[1].plot(model12_hist['loss'])
ax[1].plot(model12_hist['val_loss'])
ax[1].set_title('model 12 loss')
ax[1].set_ylabel('loss')
ax[1].set_xlabel('epoch')
ax[1].legend(['train', 'test'], loc='upper right')

plt.show()
```

    Does not look very good but learning rate seem to be set correctly

### Adding dropout at the 3th and 4th layers (between one with 256 feature map and one with 384 feature map)

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

### Model 13
plt.suptitle('Classifer 2, dropout every layer', fontsize=20)

ax[0].plot(model13_hist['acc'])
ax[0].plot(model13_hist['val_acc'])
ax[0].set_title('model 13 accuracy')
ax[0].set_ylabel('accuracy')
ax[0].set_xlabel('epoch')
ax[0].legend(['train', 'test'], loc='lower right')

ax[1].plot(model13_hist['loss'])
ax[1].plot(model13_hist['val_loss'])
ax[1].set_title('model 13 loss')
ax[1].set_ylabel('loss')
ax[1].set_xlabel('epoch')
ax[1].legend(['train', 'test'], loc='upper right')

plt.show()
```
